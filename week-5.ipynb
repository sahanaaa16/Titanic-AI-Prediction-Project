```python
data.head()
```

```python
features = ['Age']
X = data[features]
y = data.Survived
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier ()
model.fit(X,y)
```

```python
from sklearn.metrics import classification_report

predict = model.predict(X)

print(classification_report(y, predict))
```
